## Types
- Logical errors - may not cause any error msg instead the o/p produce unintended result: >, <, etc.
- Syntex errors - invalid use of rules: misspelling, paranthesis, colo etc.
- Exceptions - happens when program doesn't know how to execute code, even though there is no problem with syntax: like something mathimetically impossible -> divide something by 0, index value that doesn't exist, when python doesn't recognize variables or function names, also when we use incorrect data type.
