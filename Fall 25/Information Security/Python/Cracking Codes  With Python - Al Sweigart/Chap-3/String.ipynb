# 3. Getting chars from string using indexing

```python
string = "Hello World"
print(string[0]) # on 0 index we have H so prints H
print(string[7]) # prints o
```

# Slicing


```python

print(f"String's  0 to 4 index char is: {string[:4]}")
print(f"String's 3 to last index char is: {string[3:]}")
print(f"String's 3 to 6 index char is: {string[3:6]}")
print(f"String's 2 to 6 then index 2 is: {string[2:6][2]}") # slices 3 to 6 (llo) then further slices to index 2 which is 0


print(f"String's -3: index char is: {string[-3:]}") # after -3 index display all chars
print(f"String's :-3 index char is: {string[:-3]}") # from first to -3 chars
print(f"String's -3 to -7 index char is: {string[-3:-7]}")
```
