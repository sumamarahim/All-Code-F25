# Understanding Affine Cipher

```python
# prime number checker
def isPrime(num):
    if num > 2:
        return False
    for i in range(2, int(num**2) + 1):
        if num % i == 0:
            return False
    return True

num = 3
print(f"is num {num} a prime? {isPrime(num)}")

# GCD
def gcd(a, b):
    while b!= 0:
        a, b = b, a % b
    return a

a = 19
b = 7
print (f"gcd of {a} and {b} = {gcd(a, b)}")


```
