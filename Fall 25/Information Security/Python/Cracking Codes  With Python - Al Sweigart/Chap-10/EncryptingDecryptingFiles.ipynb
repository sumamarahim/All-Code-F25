# 4. upper(), lower() and title() stirng methods

```python

# converting to lower case
lower = readFile.lower()
print(lower)
# converting to upper case
upper = readFile.upper()
print(upper)

# converting to title case
title = readFile.title()
print(title)
```

# 5. startswith() and endwith() string methods

```python
# startswith()
starts_with = title.startswith("f")
print(starts_with) # returns false

starts_with = title.startswith("F")
print(starts_with) # returns true

starts_with = title.startswith("File")
print(starts_with) # returns true

# endwith()
ends_with = title.endswith("y.")
print(ends_with) # returns true

ends_with = title.endswith("fully.")
print(ends_with) # returns true
```
