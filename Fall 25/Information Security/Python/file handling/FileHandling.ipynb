# File Handling in python

```python
print("File Handling in python!")
```

## Opening file - Method 1 (no need to close file at end)

```python
# to open file
fileName = "test.txt"
with open(fileName, "r") as file:
    file_text = file.read()
print(file_text)
```

## Opening file - Method 2 (need to close file at end)

```python
fileName = "test.txt"
file = open(fileName, "r")
print(file.read())
# after reading the file, we need to close it.
file.close()
```
